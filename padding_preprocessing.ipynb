```
#Remove unnecessary feature
import gc
def pop(ls):
  ls=ls.replace('\t',',').split(',')
  ls.pop(0)
  ls.pop(1)
  ls.pop(3)
  return ls
```

```
linestotal = [pop(line.rstrip('\n')) for line in open('/content/EP1.01.txt')]
gc.collect()
```

```
#since it is a big data problem, fragmenting into smaller pieces because all the data cannot fit into ram.
x=[(0,300000),(300000,600000),(60000,None)]
y=1
for i in x:
  lines=linestotal[i[0]:i[1]]
  with open('/content/drive/My Drive/BCI MNIST/outfile.'+str(y), 'wb') as fp:
    pickle.dump(lines, fp)
  y+=1
  lines=None
  gc.collect() #collect garbage to free up space
```
